```python
import scrapy
from scrapy.crawler import CrawlerProcess
from twisted.internet import reactor
import logging

class ScrapingWeb(scrapy.Spider):
    name = 'sindonews'
    keyword = 'teknologi'
    start_url = [
        'https://tekno.sindonews.com/'+keyword
        ]
    custom_settings = {
        'FEED_FORMAT': 'csv',
        'FEED_URI': 'berita.csv'
        }
    def parse(self, response):
        articles = response.css('div.articles--iridescent-list')
        logging.info(f"Found {len(articles)} articles.")

        for data in articles:
            title = data.css('span.articles--iridescent-list--text-item__title-link-text::text').get()
            summary = data.css('div.articles--iridescent-list--text-item__summary::text').get()
            category = data.css('a.articles--iridescent-list--text-item__category::text').get()
            date = data.css('time.timeago::text').get()

            logging.info(f"Title: {title}, Summary: {summary}, Category: {category}, Date: {date}")

            if title and summary and category and date:
                yield {
                    'judul': title,
                    'isi': summary,
                    'kategori': category,
                    'tanggal': date,
                }
            else:
                logging.warning("One or more fields are missing data.")

# Memulai proses crawling
process = CrawlerProcess()
process.crawl(ScrapingWeb)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Membuat variabel list kosong untuk menampung hasil berita
article_result = []

# Membuat dictionary untuk menghitung jumlah artikel per kategori
category_count = {"TEKNOLOGI": 0, "OTOMOTIF": 0}
max_articles = 50  # Jumlah maksimum artikel per kategori

# Iterasi halaman (sesuaikan page_num range jika diperlukan)
for page_num in range(1, 100):  # Anda bisa meningkatkan angka ini jika diperlukan
    url = "https://www.sindonews.com/".format(page_num)
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    articles = soup.find_all('div', class_='article_list_title')

    # Iterasi artikel dalam setiap halaman
    for article in articles:
        category = article.find('div', class_='article__subtitle').text.strip()

        # Memeriksa apakah kategori adalah "Bola" atau "Otomotif"
        if category in category_count:
            if category_count[category] < max_articles:
                article_url = article.find('a', class_='article__link')['href']
                title = article.find('h3', class_='article__title').text.strip()
                date = article.find('div', class_='article__date').text.strip()

                cPage = requests.get(article_url)
                cSoup = BeautifulSoup(cPage.text, 'html.parser')

                content_paragraphs = cSoup.find('div', class_='read__content').find_all('p')
                content = " ".join([p.text.strip() for p in content_paragraphs])

                # Menggabungkan detail artikel ke dalam list hasil
                article_result.append([title, content, date, category])

                # Menambah jumlah artikel yang telah diambil dari kategori ini
                category_count[category] += 1

                # Jika sudah mencapai 50 artikel dari kedua kategori, hentikan proses
                if category_count["TEKNOLOGI"] >= max_articles and category_count["OTOMOTIF"] >= max_articles:
                    break

    # Jika sudah mencapai 50 artikel dari kedua kategori, hentikan pencarian di halaman berikutnya
    if category_count["TEKNOLOGI"] >= max_articles and category_count["OTOMOTIF"] >= max_articles:
        break
# Konversi hasil ke DataFrame
df = pd.DataFrame(article_result, columns=['title', 'content', 'date', 'category'])
```
